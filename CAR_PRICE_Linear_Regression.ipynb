```python
# Importing Libraries
import pandas as pd
import numpy as np 
from sklearn import linear_model
```

```python
#Reading Data_Set
df = pd.read_csv('car_data.csv')
df.head()
```

```python
df.tail()
```

```python
inputs = df.drop(['Car_Name','Owner','Selling_type','Present_Price'],axis='columns')
target = df.Selling_Price
target
```

```python
inputs
```

LABEL ENCODING 

```python
from sklearn.preprocessing import LabelEncoder #sklearn.preprocessing package
Numerics = LabelEncoder()
```

```python
inputs['Fuel_Type_numerical'] = Numerics.fit_transform(inputs['Fuel_Type'])
inputs['Transmission_numerical'] = Numerics.fit_transform(inputs['Transmission'])
```

```python
inputs
```

```python
inputs_new = inputs.drop(['Fuel_Type','Transmission','Selling_Price'],axis='columns')
inputs_new
```

LINEAR REGRESSION

```python
model = linear_model.LinearRegression()
```

```python
model.fit(inputs_new.values,target)
```

PREDICTION

```python
predicted = model.predict([[2014,42450,1,1]])
print(predicted)
```
